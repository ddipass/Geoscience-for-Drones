## ğŸ—ºï¸ 2. ä¸“ä¸šåœ°ç†æ•°æ®å åŠ åˆ†æ

```python
# ğŸ—ºï¸ åˆ›å»ºä¸“ä¸šåœ°ç†æ•°æ®å åŠ åœ°å›¾
print("ğŸ—ºï¸ åˆ›å»ºæµ·å—ä½ç©ºç»æµåœ°ç†æ•°æ®å åŠ åˆ†æ...")

m_geo = leafmap.Map(center=hainan_center, zoom=8, height="600px")
m_geo.add_basemap('CartoDB.Positron')  # ä½¿ç”¨æµ…è‰²åº•å›¾ä¾¿äºæ•°æ®å åŠ 

# æ·»åŠ æµ·å—çœè¡Œæ”¿è¾¹ç•Œï¼ˆæ¨¡æ‹ŸGeoJSONæ•°æ®ï¼‰
# åœ¨å®é™…åº”ç”¨ä¸­ï¼Œè¿™é‡Œä¼šä½¿ç”¨çœŸå®çš„æµ·å—çœè¾¹ç•Œæ•°æ®
hainan_geojson_url = 'https://raw.githubusercontent.com/opengeos/leafmap/master/examples/data/countries.geojson'

# è‡ªå®šä¹‰æ ·å¼
boundary_style = {
    "stroke": True,
    "color": "#FF0000",
    "weight": 3,
    "opacity": 1,
    "fill": True,
    "fillColor": "#FF0000",
    "fillOpacity": 0.1,
}

hover_style = {"fillOpacity": 0.3}

try:
    m_geo.add_geojson(
        hainan_geojson_url, 
        layer_name="æµ·å—çœè¾¹ç•Œ", 
        style=boundary_style, 
        hover_style=hover_style
    )
    print("âœ… æ·»åŠ æµ·å—çœè¡Œæ”¿è¾¹ç•Œ")
except:
    print("âš ï¸ è¾¹ç•Œæ•°æ®åŠ è½½å¤±è´¥ï¼Œä½¿ç”¨å¤‡ç”¨æ–¹æ¡ˆ")

m_geo.attribution_control = False
print("ğŸ—ºï¸ åœ°ç†æ•°æ®å åŠ åˆ†æåœ°å›¾")
m_geo
```

## ğŸ“Š 3. ä½ç©ºç»æµçƒ­åŠ›å›¾åˆ†æ

```python
# ğŸ“Š åˆ›å»ºä½ç©ºç»æµæ´»åŠ¨çƒ­åŠ›å›¾
print("ğŸ“Š åˆ›å»ºæµ·å—ä½ç©ºç»æµæ´»åŠ¨çƒ­åŠ›å›¾...")

# æ¨¡æ‹Ÿæµ·å—ä½ç©ºç»æµæ´»åŠ¨æ•°æ®
low_altitude_data = {
    'latitude': [20.0444, 18.2577, 19.5175, 19.2463, 19.6127, 18.7962, 19.1667, 18.7833],
    'longitude': [110.1989, 109.5185, 109.5809, 110.4664, 110.7539, 110.3887, 110.5833, 109.5167],
    'activity_level': [95, 88, 75, 65, 82, 58, 70, 45],
    'location': ['æµ·å£å¸‚', 'ä¸‰äºšå¸‚', 'å„‹å·å¸‚', 'ç¼æµ·å¸‚', 'æ–‡æ˜Œå¸‚', 'ä¸‡å®å¸‚', 'åšé³Œé•‡', 'äº”æŒ‡å±±å¸‚']
}

# åˆ›å»ºCSVæ•°æ®
import pandas as pd
df = pd.DataFrame(low_altitude_data)
df.to_csv('hainan_low_altitude_activity.csv', index=False)
print(f"âœ… åˆ›å»ºæ´»åŠ¨æ•°æ®æ–‡ä»¶ï¼ŒåŒ…å«{len(df)}ä¸ªç›‘æµ‹ç‚¹")

# åˆ›å»ºçƒ­åŠ›å›¾
m_heat = leafmap.Map(center=hainan_center, zoom=8, height="500px")
m_heat.add_basemap('CartoDB.DarkMatter')  # æ·±è‰²åº•å›¾çªå‡ºçƒ­åŠ›æ•ˆæœ

m_heat.add_heatmap(
    'hainan_low_altitude_activity.csv',
    latitude='latitude',
    longitude='longitude',
    value='activity_level',
    name='ä½ç©ºç»æµæ´»åŠ¨çƒ­åŠ›å›¾',
    radius=25,
    blur=15
)

# æ·»åŠ é¢œè‰²æ¡
colors = ['blue', 'cyan', 'lime', 'yellow', 'red']
m_heat.add_colorbar(colors=colors, vmin=0, vmax=100, caption='æ´»åŠ¨å¼ºåº¦æŒ‡æ•°')

m_heat.attribution_control = False
print("ğŸ“Š æµ·å—ä½ç©ºç»æµæ´»åŠ¨çƒ­åŠ›å›¾")
m_heat
```

## ğŸ¯ 4. ROIåŒºåŸŸé€‰æ‹©ä¸ä¸“é¢˜åˆ†æ

```python
# ğŸ¯ åˆ›å»ºROIé€‰æ‹©å’Œä¸“é¢˜åˆ†æ
print("ğŸ¯ åˆ›å»ºROIåŒºåŸŸé€‰æ‹©ä¸ä¸“é¢˜åˆ†æ...")

m_roi = leafmap.Map(center=hainan_center, zoom=9, height="600px")
m_roi.add_basemap('Esri.WorldImagery')

# å¯ç”¨ç»˜å›¾å·¥å…·è¿›è¡ŒROIé€‰æ‹©
m_roi.draw_control = True
m_roi.measure_control = True

print("ğŸ’¡ ä½¿ç”¨è¯´æ˜:")
print("1. ä½¿ç”¨å·¦ä¾§ç»˜å›¾å·¥å…·é€‰æ‹©æ„Ÿå…´è¶£åŒºåŸŸ(ROI)")
print("2. å¯ä»¥ç»˜åˆ¶çŸ©å½¢ã€å¤šè¾¹å½¢æˆ–åœ†å½¢åŒºåŸŸ")
print("3. ä½¿ç”¨æµ‹é‡å·¥å…·è®¡ç®—è·ç¦»å’Œé¢ç§¯")
print("4. é€‰æ‹©å®Œæˆåå¯è¿›è¡Œè¿›ä¸€æ­¥åˆ†æ")

m_roi.attribution_control = False
print("ğŸ¯ ROIé€‰æ‹©ä¸ä¸“é¢˜åˆ†æåœ°å›¾")
m_roi
```

```python
# è·å–ç”¨æˆ·ç»˜åˆ¶çš„ROIè¾¹ç•Œ
print("ğŸ“ è·å–ROIè¾¹ç•Œåæ ‡...")

# å°è¯•è·å–ç”¨æˆ·ç»˜åˆ¶çš„è¾¹ç•Œ
try:
    bbox = m_roi.user_roi_bounds()
    if bbox is None:
        # å¦‚æœç”¨æˆ·æ²¡æœ‰ç»˜åˆ¶ï¼Œä½¿ç”¨é»˜è®¤çš„æµ·å£åŒºåŸŸ
        bbox = [110.0, 19.9, 110.4, 20.2]
        print("âš ï¸ æœªæ£€æµ‹åˆ°ç”¨æˆ·ç»˜åˆ¶åŒºåŸŸï¼Œä½¿ç”¨é»˜è®¤æµ·å£åŒºåŸŸ")
    else:
        print(f"âœ… æ£€æµ‹åˆ°ç”¨æˆ·é€‰æ‹©åŒºåŸŸ: {bbox}")
except:
    bbox = [110.0, 19.9, 110.4, 20.2]
    print("âš ï¸ ROIè·å–å¤±è´¥ï¼Œä½¿ç”¨é»˜è®¤åŒºåŸŸ")

print(f"ğŸ“Š åˆ†æåŒºåŸŸè¾¹ç•Œ: {bbox}")
print(f"ğŸ“ åŒºåŸŸé¢ç§¯: çº¦{abs((bbox[2]-bbox[0])*(bbox[3]-bbox[1]))*111*111:.1f} å¹³æ–¹å…¬é‡Œ")
```

## ğŸ”„ 5. æ—¶åºå¯¹æ¯”åˆ†æï¼ˆåˆ†å±å±•ç¤ºï¼‰

```python
# ğŸ”„ åˆ›å»ºæ—¶åºå¯¹æ¯”åˆ†æ
print("ğŸ”„ åˆ›å»ºæµ·å—ä½ç©ºç»æµå‘å±•æ—¶åºå¯¹æ¯”åˆ†æ...")

# ä½¿ç”¨foliumåç«¯è¿›è¡Œåˆ†å±å¯¹æ¯”
import leafmap.foliumap as leafmap_folium

# æ¨¡æ‹Ÿ2020å¹´å’Œ2024å¹´çš„å¯¹æ¯”æ•°æ®
# åœ¨å®é™…åº”ç”¨ä¸­ï¼Œè¿™é‡Œä¼šä½¿ç”¨çœŸå®çš„å†å²å«æ˜Ÿå½±åƒ
print("ğŸ“… å‡†å¤‡2020å¹´ä¸2024å¹´å¯¹æ¯”æ•°æ®...")

try:
    m_split = leafmap_folium.Map(center=hainan_center, zoom=10)
    
    # æ¨¡æ‹Ÿä½¿ç”¨ä¸åŒçš„åº•å›¾æ¥è¡¨ç¤ºä¸åŒæ—¶æœŸ
    # å®é™…åº”ç”¨ä¸­ä¼šä½¿ç”¨å†å²å«æ˜Ÿå½±åƒ
    left_layer = 'OpenStreetMap'  # ä»£è¡¨2020å¹´çŠ¶æ€
    right_layer = 'Esri.WorldImagery'  # ä»£è¡¨2024å¹´çŠ¶æ€
    
    # åˆ›å»ºåˆ†å±å¯¹æ¯”
    m_split.split_map(
        left_layer=left_layer,
        right_layer=right_layer,
        left_label='2020å¹´ - å‘å±•åˆæœŸ',
        right_label='2024å¹´ - å¿«é€Ÿå‘å±•æœŸ'
    )
    
    print("âœ… æ—¶åºåˆ†å±å¯¹æ¯”åœ°å›¾åˆ›å»ºæˆåŠŸ")
    print("ğŸ’¡ å·¦ä¾§æ˜¾ç¤º2020å¹´çŠ¶æ€ï¼Œå³ä¾§æ˜¾ç¤º2024å¹´çŠ¶æ€")
    print("ğŸ” å¯ä»¥æ‹–åŠ¨ä¸­é—´åˆ†å‰²çº¿è¿›è¡Œå¯¹æ¯”åˆ†æ")
    
    m_split
    
except Exception as e:
    print(f"âš ï¸ åˆ†å±åŠŸèƒ½æš‚ä¸å¯ç”¨: {e}")
    print("ä½¿ç”¨å•ä¸€åœ°å›¾å±•ç¤ºæœ€æ–°çŠ¶æ€")
    
    m_current = leafmap.Map(center=hainan_center, zoom=10, height="500px")
    m_current.add_basemap('Esri.WorldImagery')
    m_current.attribution_control = False
    print("ğŸ“Š å½“å‰çŠ¶æ€åœ°å›¾")
    m_current
```
